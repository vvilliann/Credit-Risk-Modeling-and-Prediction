# Part 3: Data Cleaning and Feature Preprocessing

```python
df.head()
```

```python
df.describe()
```

```python
# Data Cleaning / Intergration
class_mapping = {label:index for index, label in enumerate(np.unique(df['term']))}
df['term'] = df['term'].map(class_mapping)

class_mapping = {label:index for index, label in enumerate(np.unique(df['grade']))}
df['grade'] = df['grade'].map(class_mapping)

class_mapping = {label:index for index, label in enumerate(np.unique(df['home_ownership']))}
df['home_ownership'] = df['home_ownership'].map(class_mapping)

class_mapping = {label:index for index, label in enumerate(np.unique(df['verification_status']))}
df['verification_status'] = df['verification_status'].map(class_mapping)

class_mapping = {label:index for index, label in enumerate(np.unique(df['purpose']))}
df['purpose'] = df['purpose'].map(class_mapping)

class_mapping = {label:index for index, label in enumerate(np.unique(df['addr_state']))}
df['addr_state'] = df['addr_state'].map(class_mapping)

class_mapping = {'Fully Paid': 0, 'Charged Off': 1}
df['loan_status'] = df['loan_status'].map(class_mapping)

df.head()
```

```python
# deal with missing value
df = df.select_dtypes(include=[np.number]).interpolate().dropna()
# drop high correlation and high variance colums
df = df.drop(['total_pymnt'], axis = 1)
df = df.drop(['total_pymnt_inv'], axis = 1)
df = df.drop(['total_rec_int'], axis = 1)
df = df.drop(['id'], axis = 1)
df = df.drop(['total_rec_prncp'], axis = 1)

df.head(10)
```

## Data Split

```python
from sklearn.model_selection import train_test_split

y_pred = df.loan_status
x_clean = df.drop(['loan_status'], axis = 1)

# keep 25% for testing
A_train, A_test, B_train, B_test = train_test_split(x_clean, y_pred, random_state=42, test_size=0.25)
```

```python
A_train.head()
```

```python
B_train.head()
```

## Data Preprocessing

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(A_train)
A_train = scaler.transform(A_train)
A_test = scaler.transform(A_test)
```

# Part 4: Model Training and Selection

## 3.1: RandomForest/ Kneighbors/ LogisticRegression

## 3.1.1 Build Models

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression

classfier_logistic = LogisticRegression()
classfier_KNN = KNeighborsClassifier()
classfier_RF = RandomForestClassifier()
```

```python
classfier_logistic.fit(A_train, B_train)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
prediction_Regression = classfier_logistic.predict(A_test)
print(classification_report(B_test,prediction_Regression))
```

```python
classfier_KNN.fit(A_train, B_train)
```

```python
prediction_KNN = classfier_KNN.predict(A_test)
print(classification_report(B_test, prediction_KNN))
```

```python
classfier_RF.fit(A_train, B_train)
```

```python
prediction_RF = classfier_RF.predict(A_test)
print(classification_report(B_test, prediction_RF))
```

## Cross Validation

```python
from sklearn import model_selection

model_names = ['Logistic Regression','KNN','Random Forest']
model_list = [classfier_logistic, classfier_KNN, classfier_RF]
count = 0

for classfier in model_list:
  cv_score = model_selection.cross_val_score(classfier, A_train, B_train, cv=5)
  print(cv_score)
  print('Model accuracy of ' + model_names[count] + ' is ' + str(cv_score.mean()))
  count += 1
```

## Other Models

```python
# SVM
from sklearn.svm import SVC

classifier_SVC = SVC()

cv_score = model_selection.cross_val_score(classifier_SVC, A_train, B_train, cv=5)
print('Model accuracy of SVM is: ' + str(cv_score.mean()))
```

```python
# Nerual Network
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(30,30,30))
mlp.fit(A_train,B_train)
```

```python
predictions = mlp.predict(A_test)
from sklearn.metrics import classification_report,confusion_matrix
print(confusion_matrix(B_test,predictions))
```

```python
print(classification_report(B_test,predictions))
```

```python
forest = RandomForestClassifier()
forest.fit(x_clean, y_pred)

importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

print("Feature importance ranking by Random Forest Model:")
for ind in range(x_clean.shape[1]):
  print("{0}:{1}".format(x_clean.columns[indices[ind]], round(importances[indices[ind]], 4)))
```

```python
# Add L1 regularization to logistic regression
# check the Coefficients for feature selection
scaler = StandardScaler()
A_l1 = scaler.fit_transform(A_with_corr)
LRmodel_l1 = LogisticRegression(penalty='l1', C=10, solver='liblinear')
LRmodel_l1.fit(A_l1, y_pred)

indices = np.argsort(abs(LRmodel_l1.coef_[0]))[::-1]

print("Logistic Regression (L1) Coefficients")
for ind in range(A_with_corr.shape[1]):
  print("{0}:{1}".format(A_with_corr.columns[indices[ind]], round(LRmodel_l1.coef_[0][indices[ind]], 4)))
```
